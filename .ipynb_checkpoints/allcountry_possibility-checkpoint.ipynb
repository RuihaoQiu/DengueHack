<b>This notebook is gathering the global data of several main factors (temperature, precipation, population and urbanization) that influence both Aedes mosquitos and the Dengue virus, and reported dengue cases; using advanced machine learning algorithm to predict the possibily of dengue break in unknown areas.</b>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import ensemble
from sklearn import datasets
from sklearn.utils import shuffle
from sklearn.metrics import mean_squared_error

%matplotlib inline

# =================================================================================
#  Import training and testing data for the area have reported dengue cases
#  Reported yes as 1, reported no as 0, on possibility column
# =================================================================================
pd.options.display.max_rows=200
df1 = pd.read_csv('data/Training_data.csv')
df1.head()
```

```python
# =================================================================================
#  Randomly take 90% of data as training data and the rest 10% as testing data
# =================================================================================

df_tr = df1.sample(frac=0.9, random_state=0)
df_te = df1.loc[~df1.index.isin(df_tr.index)]
```

```python
# =================================================================================
#  Use Gradient boosting regression method 
#  build model with training data and test with testing data 
# =================================================================================

x_tr = np.array(df_tr.loc[:,'Temperature':'Urbanization'])
y_tr = np.array(df_tr['Possibility'])


params = {'n_estimators': 200, 'max_depth': 4, 'min_samples_split': 2,
          'learning_rate': 0.01, 'loss': 'ls'}

clf = ensemble.GradientBoostingRegressor(**params)
clf.fit(x_tr, y_tr)

y_te = np.array(df_te['Possibility'])
x_te = np.array(df_te.loc[:,'Temperature':'Urbanization'])

y = clf.predict(x_te)
for i in range(len(y)):
    if y[i] >= 1:
        y[i] = 1.0
    elif y[i] <= 0:
        y[i] = 0.0

mse = mean_squared_error(y_te, clf.predict(x_te))
    
print("Predict data from the model:")
print y
print("Actual data:")
print y_te
print("MSE: %.4f" % mse)
```

```python
# =================================================================================
#  Importance of each factor
# =================================================================================
feature_importance = clf.feature_importances_
print feature_importance
```

```python
# =================================================================================
#  Import data and make predictions from the model
# =================================================================================
df2 = pd.read_csv('data/Testing_data.csv')
xx = np.array(df2.loc[:,'Temperature':'Urbanization'])

y = clf.predict(xx)

for i in range(len(y)):
    if y[i] >= 1:
        y[i] = 1.0
    elif y[i] <= 0:
        y[i] = 0.0

print y

df2['Possibility'] = y
df2
df2.sort_values(by=['Dengue cases'], ascending=[0])
```
